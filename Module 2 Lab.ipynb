```python
while True:
    # Ask for the student's last name
    last_name = input("Enter the student's last name (or 'ZZZ' to quit): ")
    if last_name == 'ZZZ':
        print("Exiting program... Goodbye!")
        break  # Exit the loop if user types ZZZ

    # Ask for the student's first name
    first_name = input("Enter the student's first name: ")

    # Ask for the student's GPA and convert to float
    gpa = float(input("Enter the student's GPA: "))

    # Decide Dean's List or Honor Roll
    if gpa >= 3.5:
        print(first_name + " " + last_name + " has made the Dean's List!")
    elif gpa >= 3.25:
        print(first_name + " " + last_name + " has made the Honor Roll!")
    else:
        print(first_name + " " last_name} did not qualify for either list.")
```
